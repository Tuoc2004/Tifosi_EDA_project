# Crawl data

```python
from facebook_scraper import get_posts,get_profile
import pandas as pd
from datetime import datetime
```

```python
FANPAGE_LINK ="tifosi.hpo"
FOLDER_PATH = "Data/"
COOKIE_PATH = "./cookies.txt"
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True,page_limit = 20, cookies=COOKIE_PATH, timeout = 10):
    print(post)
    print(len(post_list))
    post_list.append(post)
```

```python
# Initialize dataframe to scrape Facebook post
post_df_full = pd.DataFrame(columns = [])
# Start to collect Facebook post data by facebook_scraper library
for post in post_list:
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = pd.concat([post_df_full, fb_post_df], ignore_index=True)
    print(post['post_id']+' get')
```

```python
# post không có dữ liệu
null_text_rows = post_df_full[pd.isna(post_df_full['text'])]
print(len(null_text_rows))
```

```python
# post không có comment
null_comment_rows = post_df_full[post_df_full['comments'] < 0]
print(len(null_comment_rows))
```

# Get profile

```python
import datetime
df = pd.read_csv('Data/data_cleaned.csv')
total_comments = df['comments_full']
# Tạo một danh sách để lưu trữ tất cả các giá trị commenter_id
all_commenter_ids = set()

# Lặp qua mỗi bài viết trong danh sách total_comments
for post_data in total_comments:
    comments_data = eval(post_data)
    for comment in comments_data:
        commenter_id = comment.get('commenter_id')  # Lấy giá trị commenter_id từ mỗi comment
        if commenter_id is not None:
            all_commenter_ids.add(commenter_id)

# In ra tất cả các giá trị commenter_id
all_commenter_ids = list(all_commenter_ids)
all_commenter_ids
```

```python
len(all_commenter_ids)
```

```python
profile_list = []
for id in all_commenter_ids[46:100]:
        profile = get_profile(id, cookies=COOKIE_PATH,allow_extra_requests = False, sleep = 5)
        print(profile)
        profile_list.append(profile)
profile_list
```

```python
df = pd.DataFrame(profile_list)
```
